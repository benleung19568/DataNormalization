```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the Excel file
file_path = 'C:/Users/user/Desktop/PreNormDataWarf.xlsx'  # Replace with your file path
df = pd.read_excel(file_path)

# Create a Min-Max Scaler
scaler = MinMaxScaler()

# Get numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Prepare an empty DataFrame for normalized data
normalized_df = pd.DataFrame()

# Normalize each numeric column with progress tracking
total_columns = len(numeric_columns)
for i, column in enumerate(numeric_columns, start=1):
    # Fit and transform column, then flatten the array if needed
    normalized_values = scaler.fit_transform(df[[column]])
    normalized_df[column] = normalized_values.flatten()  # Flatten any (n,1) arrays
    progress = (i / total_columns) * 100
    print(f"Processing column {i}/{total_columns}: {column} - {progress:.2f}% completed")

# Preserve non-numeric columns
for col in df.select_dtypes(exclude=['number']).columns:
    normalized_df[col] = df[col]

# Prompt user for output file saving location
output_file = input("C:/Users/user/Desktop/normalized_data.xlsx (e.g., C:\\Users\\YourName\\Documents\\normalized_data.xlsx): ")
normalized_df.to_excel(output_file, index=False)

print(f"Normalized data saved to {output_file}")
```
